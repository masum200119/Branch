```python
import os
import torch
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torchvision.models import alexnet, AlexNet_Weights
from torch.optim import AdamW
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import (accuracy_score, classification_report,
                             confusion_matrix, roc_curve, auc,
                             precision_recall_curve, average_precision_score, f1_score,
                             precision_score, recall_score)
from sklearn.preprocessing import label_binarize
import numpy as np
from itertools import cycle
import random
import seaborn as sns

# Configuration
CONFIG = {
    "dataset_path": r"E:\Symposiam\CC",
    "image_size": 224,
    "batch_size": 16,
    "num_epochs": 20,
    "learning_rate": 5e-5,
    "weight_decay": 0.01,
    "dropout_rate": 0.5,  # AlexNet already uses 0.5 dropout in classifier
    "seed": 42
}

# Set random seeds for reproducibility
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(CONFIG["seed"])

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data preparation
def prepare_data():
    # Common transform without augmentation
    base_transform = transforms.Compose([
        # If images are grayscale, uncomment the next line to expand to 3 channels
        # transforms.Grayscale(num_output_channels=3),
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    # Augmentation for training
    train_transform = transforms.Compose([
        # transforms.Grayscale(num_output_channels=3),  # uncomment if inputs are grayscale
        transforms.Resize((CONFIG["image_size"], CONFIG["image_size"])),
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomRotation(30),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
        transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        transforms.RandomResizedCrop(CONFIG["image_size"], scale=(0.8, 1.0)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.1, scale=(0.02, 0.1), ratio=(0.3, 3.3))
    ])

    # Load datasets
    train_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "train"), transform=train_transform)
    val_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "val"), transform=base_transform)
    test_dataset = datasets.ImageFolder(os.path.join(CONFIG["dataset_path"], "test"), transform=base_transform)

    print("\nClass distribution:")
    print(f"Train: {np.bincount([y for _, y in train_dataset])}")
    print(f"Val: {np.bincount([y for _, y in val_dataset])}")
    print(f"Test: {np.bincount([y for _, y in test_dataset])}")

    return (
        DataLoader(train_dataset, batch_size=CONFIG["batch_size"], shuffle=True, num_workers=2, pin_memory=True),
        DataLoader(val_dataset, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=2, pin_memory=True),
        DataLoader(test_dataset, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=2, pin_memory=True),
        train_dataset.classes
    )

# Model setup (AlexNet)
def setup_model(num_classes):
    # Load pretrained AlexNet
    weights = AlexNet_Weights.IMAGENET1K_V1
    model = alexnet(weights=weights)

    # Replace final classifier layer while retaining Dropout layers
    in_features = model.classifier[6].in_features
    model.classifier[6] = nn.Linear(in_features, num_classes)

    # Freeze all layers first
    for p in model.parameters():
        p.requires_grad = False

    # Unfreeze last conv block(s) + classifier for fine-tuning
    # AlexNet features is a Sequential; unfreeze last part and full classifier
    for p in model.features[8:].parameters():  # layers after index 8 (conv5 and beyond)
        p.requires_grad = True
    for p in model.classifier.parameters():
        p.requires_grad = True

    return model.to(device)

# Training loop
def train_model(model, train_loader, val_loader):
    optimizer = AdamW(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=CONFIG["learning_rate"],
        weight_decay=CONFIG["weight_decay"]
    )
    criterion = nn.CrossEntropyLoss()
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.1, patience=3, verbose=True
    )

    history = {
        'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': [],
        'train_precision': [], 'val_precision': [], 'train_recall': [], 'val_recall': [],
        'train_f1': [], 'val_f1': [], 'learning_rates': []
    }

    for epoch in range(CONFIG["num_epochs"]):
        model.train()
        train_loss, train_preds, train_labels = 0.0, [], []

        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['num_epochs']} - Training"):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            train_labels.extend(labels.cpu().numpy())

        train_loss /= len(train_loader)
        train_acc = accuracy_score(train_labels, train_preds)
        train_precision = precision_score(train_labels, train_preds, average='weighted', zero_division=0)
        train_recall = recall_score(train_labels, train_preds, average='weighted', zero_division=0)
        train_f1 = f1_score(train_labels, train_preds, average='weighted', zero_division=0)

        # Validation
        val_loss, val_preds, val_labels = evaluate_model(model, val_loader, criterion)
        val_acc = accuracy_score(val_labels, val_preds)
        val_precision = precision_score(val_labels, val_preds, average='weighted', zero_division=0)
        val_recall = recall_score(val_labels, val_preds, average='weighted', zero_division=0)
        val_f1 = f1_score(val_labels, val_preds, average='weighted', zero_division=0)

        # Store metrics
        history['train_loss'].append(train_loss)
        history['val_loss'].append(val_loss)
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)
        history['train_precision'].append(train_precision)
        history['val_precision'].append(val_precision)
        history['train_recall'].append(train_recall)
        history['val_recall'].append(val_recall)
        history['train_f1'].append(train_f1)
        history['val_f1'].append(val_f1)
        history['learning_rates'].append(optimizer.param_groups[0]['lr'])

        scheduler.step(val_acc)

        print(f"Epoch {epoch+1}: Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")
        print(f"         Train F1: {train_f1:.4f} | Val F1: {val_f1:.4f} | LR: {optimizer.param_groups[0]['lr']:.6f}")

    return history, val_preds, val_labels

def evaluate_model(model, loader, criterion):
    model.eval()
    total_loss, preds, labels = 0.0, [], []
    with torch.no_grad():
        for inputs, labels_batch in tqdm(loader, desc="Evaluating"):
            inputs, labels_batch = inputs.to(device), labels_batch.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels_batch)
            total_loss += loss.item()
            preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            labels.extend(labels_batch.cpu().numpy())

    return total_loss / len(loader), preds, labels

def get_predictions_with_probabilities(model, loader):
    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []

    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Getting predictions with probabilities"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1)

            all_preds.extend(torch.argmax(outputs, dim=1).cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())

    return np.array(all_preds), np.array(all_labels), np.array(all_probs)

# --------- INDIVIDUAL PLOTTING (each plot saved/displayed separately) ---------
def plot_individual_results(history, class_names, val_preds, val_labels,
                            test_preds=None, test_labels=None, test_probs=None):

    def save_and_close(fig_name):
        plt.tight_layout()
        plt.savefig(fig_name, dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()

    # 1. Training & Validation Loss
    plt.figure(figsize=(8, 6))
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('1_loss_curve.png')

    # 2. Training & Validation Accuracy
    plt.figure(figsize=(8, 6))
    plt.plot(history['train_acc'], label='Training Accuracy')
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('2_accuracy_curve.png')

    # 3. Training & Validation Precision
    plt.figure(figsize=(8, 6))
    plt.plot(history['train_precision'], label='Training Precision')
    plt.plot(history['val_precision'], label='Validation Precision')
    plt.title('Training and Validation Precision')
    plt.xlabel('Epoch'); plt.ylabel('Precision'); plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('3_precision_curve.png')

    # 4. Training & Validation Recall
    plt.figure(figsize=(8, 6))
    plt.plot(history['train_recall'], label='Training Recall')
    plt.plot(history['val_recall'], label='Validation Recall')
    plt.title('Training and Validation Recall')
    plt.xlabel('Epoch'); plt.ylabel('Recall'); plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('4_recall_curve.png')

    # 5. Training & Validation F1
    plt.figure(figsize=(8, 6))
    plt.plot(history['train_f1'], label='Training F1')
    plt.plot(history['val_f1'], label='Validation F1')
    plt.title('Training and Validation F1-Score')
    plt.xlabel('Epoch'); plt.ylabel('F1'); plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('5_f1_curve.png')

    # 6. Learning Rate Schedule
    plt.figure(figsize=(8, 6))
    plt.plot(history['learning_rates'])
    plt.title('Learning Rate Schedule')
    plt.xlabel('Epoch'); plt.ylabel('Learning Rate'); plt.yscale('log'); plt.grid(True, alpha=0.3)
    save_and_close('6_lr_schedule.png')

    # 7. Validation Confusion Matrix
    cm_val = confusion_matrix(val_labels, val_preds)
    plt.figure(figsize=(7, 6))
    sns.heatmap(cm_val, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Validation Confusion Matrix'); plt.xlabel('Predicted'); plt.ylabel('Actual')
    save_and_close('7_val_confusion_matrix.png')

    # 8. Test Confusion Matrix (if available)
    if test_preds is not None and test_labels is not None:
        cm_test = confusion_matrix(test_labels, test_preds)
        plt.figure(figsize=(7, 6))
        sns.heatmap(cm_test, annot=True, fmt='d', cmap='Greens',
                    xticklabels=class_names, yticklabels=class_names)
        plt.title('Test Confusion Matrix'); plt.xlabel('Predicted'); plt.ylabel('Actual')
        save_and_close('8_test_confusion_matrix.png')

    # 9. ROC Curves (if probabilities available)
    if test_probs is not None and test_labels is not None:
        n_classes = len(class_names)
        test_labels_bin = label_binarize(test_labels, classes=range(n_classes))

        plt.figure(figsize=(8, 6))
        if n_classes == 2:
            fpr, tpr, _ = roc_curve(test_labels_bin[:, 1], test_probs[:, 1])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f'ROC (AUC={roc_auc:.2f})')
        else:
            for i in range(n_classes):
                fpr, tpr, _ = roc_curve(test_labels_bin[:, i], test_probs[:, i])
                roc_auc = auc(fpr, tpr)
                plt.plot(fpr, tpr, lw=2, label=f'{class_names[i]} (AUC={roc_auc:.2f})')

        plt.plot([0, 1], [0, 1], 'k--')
        plt.title('ROC Curves'); plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate')
        plt.legend(loc='lower right'); plt.grid(True, alpha=0.3)
        save_and_close('9_roc_curves.png')

    # 10. Precision-Recall Curves (if probabilities available)
    if test_probs is not None and test_labels is not None:
        n_classes = len(class_names)
        test_labels_bin = label_binarize(test_labels, classes=range(n_classes))

        plt.figure(figsize=(8, 6))
        if n_classes == 2:
            precision, recall, _ = precision_recall_curve(test_labels_bin[:, 1], test_probs[:, 1])
            ap = average_precision_score(test_labels_bin[:, 1], test_probs[:, 1])
            plt.plot(recall, precision, label=f'AP={ap:.2f}')
        else:
            for i in range(n_classes):
                precision, recall, _ = precision_recall_curve(test_labels_bin[:, i], test_probs[:, i])
                ap = average_precision_score(test_labels_bin[:, i], test_probs[:, i])
                plt.plot(recall, precision, lw=2, label=f'{class_names[i]} (AP={ap:.2f})')

        plt.title('Precision-Recall Curves'); plt.xlabel('Recall'); plt.ylabel('Precision')
        plt.legend(); plt.grid(True, alpha=0.3)
        save_and_close('10_pr_curves.png')

    # 11. Class-wise Performance Metrics (bar chart)
    if test_preds is not None and test_labels is not None:
        report = classification_report(test_labels, test_preds, target_names=class_names, output_dict=True)
        metrics = ['precision', 'recall', 'f1-score']
        class_metrics = {m: [report[c][m] for c in class_names] for m in metrics}

        x = np.arange(len(class_names)); width = 0.25
        plt.figure(figsize=(10, 6))
        for i, m in enumerate(metrics):
            plt.bar(x + i * width, class_metrics[m], width, label=m.capitalize())
        plt.xlabel('Classes'); plt.ylabel('Score'); plt.title('Class-wise Performance Metrics')
        plt.xticks(x + width, class_names, rotation=45); plt.legend(); plt.grid(True, alpha=0.3)
        save_and_close('11_classwise_metrics.png')

    # 12. Training Progress: Loss & Error
    epochs = range(1, len(history['train_loss']) + 1)
    plt.figure(figsize=(8, 6))
    plt.plot(epochs, history['train_loss'], '-', label='Training Loss')
    plt.plot(epochs, history['val_loss'], '-', label='Validation Loss')
    plt.plot(epochs, [1 - acc for acc in history['train_acc']], '--', label='Training Error')
    plt.plot(epochs, [1 - acc for acc in history['val_acc']], '--', label='Validation Error')
    plt.title('Training Progress: Loss and Error'); plt.xlabel('Epoch'); plt.ylabel('Loss / Error')
    plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('12_progress_loss_error.png')

    # 13. Validation Accuracy vs Loss (scatter)
    plt.figure(figsize=(8, 6))
    sc = plt.scatter(history['val_loss'], history['val_acc'], c=range(len(history['val_loss'])), s=50)
    plt.xlabel('Validation Loss'); plt.ylabel('Validation Accuracy'); plt.title('Validation Accuracy vs Loss')
    cbar = plt.colorbar(sc); cbar.set_label('Epoch')
    plt.grid(True, alpha=0.3)
    save_and_close('13_val_acc_vs_loss.png')

    # 14. Learning Rate vs Validation Accuracy
    plt.figure(figsize=(8, 6))
    plt.semilogx(history['learning_rates'], history['val_acc'], 'o-')
    plt.xlabel('Learning Rate'); plt.ylabel('Validation Accuracy'); plt.title('Learning Rate vs Validation Accuracy')
    plt.grid(True, alpha=0.3)
    save_and_close('14_lr_vs_val_acc.png')

    # 15. Training Loss Reduction Rate
    loss_reduction = [history['train_loss'][i-1] - history['train_loss'][i] for i in range(1, len(history['train_loss']))]
    plt.figure(figsize=(8, 6))
    plt.plot(range(2, len(history['train_loss']) + 1), loss_reduction)
    plt.xlabel('Epoch'); plt.ylabel('Loss Reduction'); plt.title('Training Loss Reduction Rate')
    plt.axhline(y=0, linestyle='--', alpha=0.5); plt.grid(True, alpha=0.3)
    save_and_close('15_loss_reduction_rate.png')

    # 16. Overfitting Analysis (Train - Val accuracy gap)
    overfitting_gap = [tr - va for tr, va in zip(history['train_acc'], history['val_acc'])]
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(overfitting_gap) + 1), overfitting_gap)
    plt.xlabel('Epoch'); plt.ylabel('Accuracy Gap (Train - Val)'); plt.title('Overfitting Analysis')
    plt.axhline(y=0, linestyle='--', alpha=0.5); plt.grid(True, alpha=0.3)
    save_and_close('16_overfitting_gap.png')

    # 17. Metric Correlation Matrix
    metrics_df = {
        'Train Loss': history['train_loss'],
        'Val Loss': history['val_loss'],
        'Train Acc': history['train_acc'],
        'Val Acc': history['val_acc'],
        'Train F1': history['train_f1'],
        'Val F1': history['val_f1']
    }
    mat = np.corrcoef([metrics_df[k] for k in metrics_df.keys()])
    plt.figure(figsize=(8, 6))
    sns.heatmap(mat, annot=True, cmap='coolwarm', center=0,
                xticklabels=list(metrics_df.keys()), yticklabels=list(metrics_df.keys()))
    plt.title('Metric Correlation Matrix')
    save_and_close('17_metric_correlation.png')

    # 18. Per-class Accuracy (if available)
    if test_preds is not None and test_labels is not None:
        per_class_acc = []
        for i in range(len(class_names)):
            mask = (np.array(test_labels) == i)
            per_class_acc.append(accuracy_score(np.array(test_labels)[mask], np.array(test_preds)[mask]) if mask.sum() > 0 else 0.0)

        plt.figure(figsize=(10, 6))
        plt.bar(range(len(class_names)), per_class_acc)
        plt.xlabel('Classes'); plt.ylabel('Accuracy'); plt.title('Per-class Accuracy')
        plt.xticks(range(len(class_names)), class_names, rotation=45); plt.grid(True, alpha=0.3)
        save_and_close('18_per_class_accuracy.png')

    # 19. Training Stability (Rolling Std of Loss)
    window = 3
    rolling_train_std, rolling_val_std = [], []
    for i in range(len(history['train_loss'])):
        s = max(0, i - window + 1); e = i + 1
        rolling_train_std.append(np.std(history['train_loss'][s:e]))
        rolling_val_std.append(np.std(history['val_loss'][s:e]))

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(rolling_train_std) + 1), rolling_train_std, label='Train Loss Std')
    plt.plot(range(1, len(rolling_val_std) + 1), rolling_val_std, label='Val Loss Std')
    plt.xlabel('Epoch'); plt.ylabel('Loss Std Dev'); plt.title('Training Stability (Rolling Std)')
    plt.legend(); plt.grid(True, alpha=0.3)
    save_and_close('19_training_stability.png')

    # 20. Overall Performance Trend (mean of Val Acc & Val F1)
    performance_score = [(acc + f1) / 2 for acc, f1 in zip(history['val_acc'], history['val_f1'])]
    best_epoch = int(np.argmax(performance_score) + 1)
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(performance_score) + 1), performance_score)
    plt.axvline(x=best_epoch, linestyle='--', alpha=0.7)
    plt.text(best_epoch, max(performance_score), f'Best: {best_epoch}', rotation=90, va='bottom')
    plt.xlabel('Epoch'); plt.ylabel('Combined Performance Score'); plt.title('Overall Performance Trend')
    plt.grid(True, alpha=0.3)
    save_and_close('20_overall_performance_trend.png')

# Main execution
if __name__ == "__main__":
    # Prepare data
    train_loader, val_loader, test_loader, class_names = prepare_data()

    # Setup model
    model = setup_model(len(class_names))
    print(f"\nModel has {sum(p.numel() for p in model.parameters() if p.requires_grad)} trainable parameters")

    # Train model
    print("\nStarting training...")
    history, val_preds, val_labels = train_model(model, train_loader, val_loader)

    # Evaluate on test set
    print("\nEvaluating on test set...")
    test_preds, test_labels, test_probs = get_predictions_with_probabilities(model, test_loader)

    # Plot individual results (each saved separately)
    print("\nGenerating individual visualizations...")
    plot_individual_results(history, class_names, val_preds, val_labels, test_preds, test_labels, test_probs)

    # Save model
    torch.save(model.state_dict(), 'final_alexnet_model.pth')
    print("\nTraining completed and model saved as final_alexnet_model.pth!")
```
